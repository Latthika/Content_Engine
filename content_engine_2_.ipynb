```python
%%writefile app.py
import streamlit as st
from sentence_transformers import SentenceTransformer
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import chromadb
import PyPDF2

# Function to parse PDF documents and extract text
def parse_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
    return text

# Function to generate insights using a GPT-2 model
def generate_insights(prompt, model, tokenizer):
    inputs = tokenizer.encode(prompt, return_tensors="pt")
    outputs = model.generate(inputs, max_length=150, num_return_sequences=1)
    insight = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Ensure the insight does not repeat the prompt
    insight = insight.replace(prompt, '').strip()
    return insight

# Load models
embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
llm_model = GPT2LMHeadModel.from_pretrained("gpt2")

# Initialize ChromaDB and collection
client = chromadb.Client()

# Check if collection exists, otherwise create it
try:
    collection = client.get_collection("pdf_documents")
except ValueError:
    collection = client.create_collection("pdf_documents")

# Paths to the three PDF documents
pdf_paths = [
    "/content/alphabet.pdf",
    "/content/tesla.pdf",
    "/content/uber.pdf"
]

# Parse and embed documents
for i, pdf_path in enumerate(pdf_paths):
    document_text = parse_pdf(pdf_path)
    document_embedding = embedding_model.encode(document_text).tolist()
    # Adding documents to the collection with unique IDs
    collection.add(
        documents=[document_text],
        embeddings=[document_embedding],
        ids=[f"document_{i}"]
    )

# Main function for Streamlit app
def main():
    st.markdown("<h1 style='text-align: center; color: red;'>Content Engine Bot</h1>", unsafe_allow_html=True)
    st.markdown("<h2 style='text-align: center; color: blue;'>Dedicated to Uber, Tesla, and Alphabet Companies</h2>", unsafe_allow_html=True)
    st.write("Analyze and compare multiple PDF documents.")

    if 'chat_history' not in st.session_state:
        st.session_state.chat_history = []

    def add_message(user, message):
        st.session_state.chat_history.append({"user": user, "message": message})

    def display_chat():
        for chat in st.session_state.chat_history:
            if chat["user"] == "User":
                st.markdown(f"<div style='text-align: right; color: blue;'>{chat['message']}</div>", unsafe_allow_html=True)
            else:
                st.markdown(f"<div style='text-align: left; color: green;'>{chat['message']}</div>", unsafe_allow_html=True)

    # Display chat history
    display_chat()

    query = st.text_input("Message:", key="user_query")
    if st.button("Send"):
        if query:
            add_message("User", query)
            insight = generate_insights(query, llm_model, tokenizer)
            add_message("Bot", insight)
            st.experimental_rerun()

if __name__ == "__main__":
    main()
```
